```python
listing = df1.groupby("condition")

listing.get_group("Listing Expired")     #convert into 0 minutes
```

```python
df1.replace(to_replace= "Listing Expired", value = "0 minutes", inplace=True)
```

```python
df1[['condition', 'Duration','left']] = df1.condition.str.split(" ", expand = True)
```

```python
df1.drop("left",axis=1,inplace=True)
```

```python
df1.head()
```

```python
df1.replace(to_replace=" usa",value="usa",inplace=True)
```

```python
df1["condition"]= df1["condition"].astype(int)
```

```python
df1.info()
```

```python
df1.index
```

```python
for i in df1.index:
    print(df1['Duration'].iloc[i])
```

```python
#for loop - time calculation

for index, row in df1.iterrows():
    if row['Duration'] == "hours":
        df1.at[index, 'condition'] = row['condition'] * 60
    elif row['Duration'] == "days":
        df1.at[index, 'condition'] = row['condition'] * 1440
    else:
        df1.at[index, 'condition'] = 0
```

The code iterates over each row of a Pandas dataframe df1 using df1.iterrows() method. For each row, the code checks the value in the Duration column. If the value is "hours", it multiplies the value in the condition column with 60 and updates the condition column of the same row with the result. If the value is "days", it multiplies the value in the condition column with 1440 and updates the condition column of the same row with the result. If the value is neither "hours" nor "days", it sets the condition column of the same row to 0.
